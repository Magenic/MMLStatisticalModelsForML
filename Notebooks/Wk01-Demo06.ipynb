```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import gc
import math

np.random.seed(1)
tf.keras.backend.set_floatx('float64')
tf.random.set_seed(5)
tf.random.shuffle(tf.range(5))

samples, timesteps, features = 3, 10, 2
inputs = np.random.random([samples, timesteps, features]).astype(np.float64)

model = tf.keras.models.Sequential()

output = model(inputs)

print(output)
# Given inputs are radius of a circle, convert them to the area of a circle
model.add(tf.keras.layers.Lambda(lambda x: (math.pi * x) ** 2, input_shape=(timesteps, features)))
output = model(inputs)

print(output)
```
